```
import pandas as pd

df = pd.read_csv('/Winter_Fashion_Trends_Dataset.csv')

print("Первые 5 строк:")
print(df.head())
```

```
print("Первые 5 строк DataFrame:")
display(df.head())

print("\nПоследние 5 строк DataFrame:")
display(df.tail())
```

```
df.info()
```

```
for col in ['Brand', 'Category', 'Color', 'Material', 'Style', 'Gender', 'Season', 'Trend_Status']:
    if col in df.columns and df[col].dtype == 'object':
        df[col] = df[col].astype('category')

print("Типы данных после преобразования:")
df.info()
```

```
print("Количество пропущенных значений в каждом столбце:")
missing_values = df.isnull().sum()
display(missing_values[missing_values > 0])

if missing_values.sum() == 0:
    print("\nПропущенных значений не обнаружено.")
else:
    print("\nОбщее количество пропущенных значений:", missing_values.sum())
```

```
num_duplicates = df.duplicated().sum()

if num_duplicates == 0:
    print("Дубликатов нет")
else:
    print(f"Обнаружено {num_duplicates} дубликатов")
```

```
import numpy as np
from scipy.stats import zscore


numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

print("Анализ выбросов по Z-оценке:")
outlier_indices = set()

for col in numerical_cols:

    if col == 'ID':
        continue

    z_scores = np.abs(zscore(df[col]))
    threshold = 3
    outliers = df[z_scores > threshold]

    if not outliers.empty:
        print(f"\nОбнаружены выбросы в столбце '{col}' (Z-оценка > {threshold}):")
        display(outliers[[col]])
        outlier_indices.update(outliers.index)
    else:
        print(f"\nВыбросы в столбце '{col}' (Z-оценка > {threshold}) не обнаружены.")

if not outlier_indices:
    print("\nВ целом, выбросов по Z-оценке (>3) не обнаружено ни в одном числовом столбце.")
else:
    print(f"\nВсего обнаружено {len(outlier_indices)} уникальных строк с выбросами (Z-оценка > {threshold}).")
    print("Эти строки содержат выбросы хотя бы в одном из числовых столбцов.")
```

```
import matplotlib.pyplot as plt
import seaborn as sns

print("Визуализация выбросов с помощью Boxplotов:")


for col in numerical_cols:

    if col == 'ID':
        continue

    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot для столбца: {col}')
    plt.ylabel(col)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

```
print("Описательная статистика для числовых столбцов:")
display(df.describe())
```

```
print("Уникальные значения и их частоты для каждого столбца:")
for col in df.columns:
    print(f"\n----- Столбец: '{col}' -----")
    display(df[col].value_counts())
    print(f"Количество уникальных значений: {df[col].nunique()}")
```

```
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


numerical_cols_for_corr = df.select_dtypes(include=np.number).columns.tolist()

if 'ID' in numerical_cols_for_corr:
    numerical_cols_for_corr.remove('ID')

if not numerical_cols_for_corr:
    print("В DataFrame нет числовых столбцов, подходящих для корреляционного анализа")
else:
    print("Матрица корреляций для числовых столбцов:")
    correlation_matrix = df[numerical_cols_for_corr].corr()
    display(correlation_matrix)


    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Тепловая карта корреляционной матрицы')
    plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

print("Визуализация распределений с помощью гистограмм:")

for col in numerical_cols:

    if col == 'ID':
        continue

    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Гистограмма распределения для столбца: {col}')
    plt.xlabel(col)
    plt.ylabel('Частота')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns


avg_price_by_season = df.groupby('Season')['Price(USD)'].mean().reset_index()


avg_price_by_season['Season_Order'] = avg_price_by_season['Season'].apply(lambda x: int(x.split(' ')[1]))
avg_price_by_season = avg_price_by_season.sort_values(by='Season_Order')

plt.figure(figsize=(10, 6))
sns.lineplot(x='Season', y='Price(USD)', data=avg_price_by_season, marker='o')
plt.title('Средняя цена по сезонам')
plt.xlabel('Сезон')
plt.ylabel('Средняя цена (USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(12, 6))
sns.barplot(x='Brand', y='Price(USD)', data=df, errorbar=None)
plt.title('Средняя цена по брендам')
plt.xlabel('Бренд')
plt.ylabel('Средняя цена (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Price(USD)', data=df, errorbar=None)
plt.title('Средняя цена по категориям')
plt.xlabel('Категория')
plt.ylabel('Средняя цена (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Количество товаров по полу')
plt.xlabel('Пол')
plt.ylabel('Количество')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```
print("1. Средняя цена и популярность по брендам:")
brand_summary = df.groupby('Brand')[['Price(USD)', 'Popularity_Score']].mean()
display(brand_summary.sort_values(by='Price(USD)', ascending=False))

print("\n2. Количество товаров по категории и полу:")
gender_category_count = df.groupby(['Gender', 'Category']).size().unstack(fill_value=0)
display(gender_category_count)

print("\n3. Общая сумма продаж по статусу тренда:")
trend_total_price = df.groupby('Trend_Status')['Price(USD)'].sum().reset_index()
display(trend_total_price.sort_values(by='Price(USD)', ascending=False))
```

```
print("Сводная таблица: Средняя оценка покупателей по материалу и стилю:")
pivot_table_customer_rating = df.pivot_table(
    values='Customer_Rating',
    index='Material',
    columns='Style',
    aggfunc='mean',
    fill_value=0
)
display(pivot_table_customer_rating)
```
